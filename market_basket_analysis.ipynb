```python
import pandas as pd
import matplotlib.pyplot as plt

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
from google.colab import files

uploaded = files.upload()

df = pd.read_csv("basket_supermercado_1000.csv")

df.head()

print(df.shape)

df.info()
```

```python
df.columns

df = df.drop(columns=["transacao"])

df.head()

frequencia = df.sum().sort_values(ascending=False)

frequencia

plt.figure(figsize=(12,6))

frequencia.plot(kind='bar')

plt.title("Produtos mais frequentes")
plt.xlabel("Produtos")
plt.ylabel("Quantidade de compras")

plt.show()
```

```python
itemsets = apriori(
    df,
    min_support=0.10,
    use_colnames=True
)

itemsets.head()

itemsets.sort_values(
    by="support",
    ascending=False
)
```

```python
rules = association_rules(
    itemsets,
    metric="confidence",
    min_threshold=0.5
)

rules.head()

rules = rules[
    [
        "antecedents",
        "consequents",
        "support",
        "confidence",
        "lift"
    ]
]

rules

rules.sort_values(
    by="lift",
    ascending=False
)
```

```python
regras_fortes = rules[
    (rules["confidence"] >= 0.80) &
    (rules["lift"] > 1)
]

regras_fortes
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    rules["support"],
    rules["confidence"]
)

plt.xlabel("Suporte")
plt.ylabel("Confiança")
plt.title("Regras de Associação")

plt.show()
```

```python
for _, regra in regras_fortes.iterrows():

    print("SE:", list(regra["antecedents"]))

    print("ENTÃO:", list(regra["consequents"]))

    print(f"Suporte: {regra['support']:.3f}")

    print(f"Confiança: {regra['confidence']:.3f}")

    print(f"Lift: {regra['lift']:.3f}")

    print("-"*60)
```

```python
for index, regra in regras_fortes.iterrows():

    print(
        f"{list(regra['antecedents'])} "
        f"-> "
        f"{list(regra['consequents'])}"
    )

    print(
        f"Suporte: {regra['support']:.2f} | "
        f"Confiança: {regra['confidence']:.2f} | "
        f"Lift: {regra['lift']:.2f}"
    )

    print("-"*50)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    rules["support"],
    rules["confidence"],
    s=rules["lift"]*100
)

plt.xlabel("Suporte")
plt.ylabel("Confiança")

plt.title("Regras de Associação - Apriori")

plt.show()
```
